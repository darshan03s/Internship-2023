```python
converting multi d array to 1d array
```

```python
import numpy as np

multi_darray = np.array([[1, 2, 3], [4, 5, 6]])

one_darray = multi_darray.flatten()

print("2D Array:")
print(multi_darray)

print("\n1D Array (using flatten):")
print(one_darray)
```

```python
concatinating two arrays wrt axis
```

```python
import numpy as np

array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])

concatenated_axis0 = np.concatenate((array1, array2), axis=0)

concatenated_axis1 = np.concatenate((array1, array2), axis=1)

print("Array 1:")
print(array1)

print("\nArray 2:")
print(array2)

print("\nConcatenated along axis 0 (rows):")
print(concatenated_axis0)

print("\nConcatenated along axis 1 (columns):")
print(concatenated_axis1)
```

```python
import matplotlib.pyplot as plt

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]  

# Create a horizontal bar chart
plt.barh(languages, popularity, color='skyblue') 

# Add labels and a title
plt.xlabel('Popularity (%)')
plt.ylabel('Programming Languages')
plt.title('Popularity of Programming Languages')

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(languages, popularity, color='skyblue') 

plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')

plt.xticks(rotation=45)  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
genders = ['Men', 'Women']

means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]

bar_width = 0.35

x_men = np.arange(len(groups))
x_women = x_men + bar_width

plt.bar(x_men, means_men, width=bar_width, label='Men', color='blue')
plt.bar(x_women, means_women, width=bar_width, label='Women', color='red')

plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')
plt.xticks(x_men + bar_width / 2, groups)
plt.legend()

plt.show()
```
