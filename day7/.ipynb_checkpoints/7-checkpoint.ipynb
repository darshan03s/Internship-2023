```python
create a dataframe from a dictionary and display it
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'San Francisco', 'Los Angeles']
}

df = pd.DataFrame(data)

print(df)
```

```python
program to create and display a dataframe from a specified dictionary data which has the index labels
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'San Francisco', 'Los Angeles']
}

index_labels = ['Person1', 'Person2', 'Person3']

df = pd.DataFrame(data, index=index_labels)

print(df)
```

```python
program to get the first 3 rows of a given dataframe
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 28, 22],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Miami']
}

df = pd.DataFrame(data)

first_3_rows = df.head(3)

print(first_3_rows)
```

```python
program to select the 'name' and 'score ' columns from the following dataframe
exam_data = {'name': [],'score': [], 'attempts':[], 'qualify':[]}
```

```python
import pandas as pd

exam_data = {'name': ['Alice', 'Bob', 'Charlie'],
             'score': [85, 92, 78],
             'attempts': [1, 2, 3],
             'qualify': ['Yes', 'Yes', 'No']}

df = pd.DataFrame(exam_data)

selected_columns = df[['name', 'score']]

print(selected_columns)
```
